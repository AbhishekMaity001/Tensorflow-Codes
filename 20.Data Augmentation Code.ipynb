# Data Augmentation using Tensorflow and keras

In this code i have taken 1 image and did augmentaten on that and saved back to the memory

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import tensorflow as tf
```

```python
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
datagen = ImageDataGenerator(rescale=1/255,
                            rotation_range=0.4,
                            width_shift_range=0.4,
                            height_shift_range=0.4,
                            shear_range=0.4,
                            zoom_range=0.4,
                            horizontal_flip=True,
                            vertical_flip=True,channel_shift_range=3.0,
                            fill_mode='nearest')

img = image.load_img('C:\\Users\\Abhishek Maity\\Pictures\\Saved Pictures\\couple-dogs.jpg',target_size=(200,200))
x = image.img_to_array(img)
x = np.expand_dims(x,axis=0)
#x.shape # converting into 4D array

i=0
for batch in datagen.flow(x,batch_size=1,save_to_dir='augmented',save_prefix='dog',save_format='png'):
    i += 1
    if i>100:
        break # if not put the limit then it will generate an indefinite loop
```
